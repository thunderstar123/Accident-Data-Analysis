**Background and History**

Reinforcement Learning (RL) is a branch of machine learning where an agent learns to make decisions by interacting with an environment. RL is inspired by behavioral psychology and the idea of trial-and-error learning, where agents are rewarded or punished for their actions based on the outcomes.

Key historical milestones include:

1950s: Dynamic programming by Richard Bellman.

1970s: Concept of temporal-difference (TD) learning.

1989: Introduction of Q-learning by Chris Watkins.

1998: Sutton and Barto published the seminal book "Reinforcement Learning: An Introduction."

Basic Concept

Reinforcement Learning revolves around the interaction between an agent and an environment. The agent takes actions in the environment, receives feedback in the form of rewards, and updates its policy (strategy) to maximize cumulative rewards over time.

Key Components of RL

Agent: The decision-maker that interacts with the environment.

Environment: The external system that the agent operates in. The environment provides feedback based on the agent's actions.

State (S): The current situation of the agent in the environment.

Action (A): The moves or decisions the agent can make.

Reward (R): Feedback from the environment based on the agent's action.

Policy (π): A strategy or rule that the agent follows to decide which action to take in a given state.

Value Function (V): A function that estimates the future reward that can be obtained from a particular state.

Q-Function (Q): A function that estimates the value of taking a specific action in a given state.

### Mathematics

1. Markov Decision Process (MDP):
An MDP is a framework that lets us set up this problem mathematically:
* States (S): All possible situations the agent can be in, like “character in front of a monster.”
* Actions (A): Choices available to the agent in each state, like “attack” or “run.”
* Rewards (R): Feedback on each action, e.g., +10 points for collecting a coin or -5 for falling into a pit.
* Transitions (P): Probabilities of moving from one state to another after an action, e.g., a high chance of moving to “next level” if you defeat a monster.
Together, these elements form an MDP, which describes how the agent moves, earns rewards, and interacts with the environment.

2. Policy: The Agent’s Strategy
The policy is the agent’s strategy or decision-making rule. It’s like a rulebook that tells the agent, “In this situation, do this action.” Mathematically, we write the policy as 𝜋 π, which tells us the probability of choosing a certain action in a given state:

Policy (𝜋): A function mapping states to actions. If the policy says “always attack monsters when nearby,” then 𝜋(attack ∣ near monster) = 1.

3. Value Functions: Measuring How Good States and Actions Are

To decide if an action or a situation is good, we need to know the expected reward:

* Value of a State (V): How good a state is in terms
of the reward we expect to earn from it over time. For example, if entering a “safe area” state usually leads to high rewards, V(safe area) will be high.
* Value of an Action (Q): How good a specific action is in a specific state. If “attacking a monster” in “near monster” state usually leads to rewards, Q(near monster,attack) will be high.

These values help the agent choose actions that lead to the best rewards over time.

4. The Bellman Equations: Breaking Down the Reward Calculation
The Bellman equations help us break down the expected reward from any state or action by considering:

* The immediate reward from the action.
* The value of the resulting state.

One can think of it like planning moves in a game: each action gives a reward now plus the potential rewards from future actions. The Bellman equations capture this “one-step look-ahead” logic.

In this article, we are going to demonstrate how to implement a basic Reinforcement Learning algorithm which is called the Q-Learning technique. In this demonstration, we attempt to teach a bot to reach its destination using the Q-Learning technique.

```python
#Step 1: Importing the required libraries

import numpy as np
import pylab as pl
import networkx as nx
```

```python
#Step 2: Defining and visualising the graph

edges = [(0, 1), (1, 5), (5, 6), (5, 4), (1, 2),
		(1, 3), (9, 10), (2, 4), (0, 6), (6, 7),
		(8, 9), (7, 8), (1, 7), (3, 9)]

goal = 10
G = nx.Graph()
G.add_edges_from(edges)
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
pl.show()
```

```python
#Step 3: Defining the reward the system for the bot

MATRIX_SIZE = 11
M = np.matrix(np.ones(shape =(MATRIX_SIZE, MATRIX_SIZE)))
M *= -1

for point in edges:
	print(point)
	if point[1] == goal:
		M[point] = 100
	else:
		M[point] = 0

	if point[0] == goal:
		M[point[::-1]] = 100
	else:
		M[point[::-1]]= 0
		# reverse of point

M[goal, goal]= 100
print(M)
# add goal point round trip
```

```python
#Step 4: Defining some utility functions to be used in the training

Q = np.matrix(np.zeros([MATRIX_SIZE, MATRIX_SIZE]))

gamma = 0.75
# learning parameter
initial_state = 1

# Determines the available actions for a given state
def available_actions(state):
	current_state_row = M[state, ]
	available_action = np.where(current_state_row >= 0)[1]
	return available_action

available_action = available_actions(initial_state)

# Chooses one of the available actions at random
Q = np.matrix(np.zeros([MATRIX_SIZE, MATRIX_SIZE]))

gamma = 0.75
# learning parameter
initial_state = 1

# Determines the available actions for a given state
def available_actions(state):
	current_state_row = M[state, ]
	available_action = np.where(current_state_row >= 0)[1]
	return available_action

available_action = available_actions(initial_state)

# Chooses one of the available actions at random
def sample_next_action(available_actions_range):
	next_action = int(np.random.choice(available_action, 1))
	return next_action

action = sample_next_action(available_action)

def update(current_state, action, gamma):

  max_index = np.where(Q[action, ] == np.max(Q[action, ]))[1]
  if max_index.shape[0] > 1:
  	max_index = int(np.random.choice(max_index, size = 1))
  else:
	  max_index = int(max_index)
  max_value = Q[action, max_index]
  Q[current_state, action] = M[current_state, action] + gamma * max_value
  if (np.max(Q) > 0):
	  return(np.sum(Q / np.max(Q)*100))
  else:
  	return (0)
# Updates the Q-Matrix according to the path chosen

update(initial_state, action, gamma)
```

```python
#Step 5: Training and evaluating the bot using the Q-Matrix

scores = []
for i in range(1000):
	current_state = np.random.randint(0, int(Q.shape[0]))
	available_action = available_actions(current_state)
	action = sample_next_action(available_action)
	score = update(current_state, action, gamma)
	scores.append(score)

# print("Trained Q matrix:")
# print(Q / np.max(Q)*100)
# You can uncomment the above two lines to view the trained Q matrix

# Testing
current_state = 0
steps = [current_state]

while current_state != 10:

	next_step_index = np.where(Q[current_state, ] == np.max(Q[current_state, ]))[1]
	if next_step_index.shape[0] > 1:
		next_step_index = int(np.random.choice(next_step_index, size = 1))
	else:
		next_step_index = int(next_step_index)
	steps.append(next_step_index)
	current_state = next_step_index

print("Most efficient path:")
print(steps)

pl.plot(scores)
pl.xlabel('No of iterations')
pl.ylabel('Reward gained')
pl.show()
```

```python
#Step 7: Defining some utility functions for the training process

Q = np.matrix(np.zeros([MATRIX_SIZE, MATRIX_SIZE]))
env_police = np.matrix(np.zeros([MATRIX_SIZE, MATRIX_SIZE]))
env_drugs = np.matrix(np.zeros([MATRIX_SIZE, MATRIX_SIZE]))
initial_state = 1

# Same as above
def available_actions(state):
	current_state_row = M[state, ]
	av_action = np.where(current_state_row >= 0)[1]
	return av_action

# Same as above
def sample_next_action(available_actions_range):
	next_action = int(np.random.choice(available_action, 1))
	return next_action

# Exploring the environment
def collect_environmental_data(action):
	found = []
	if action in police:
		found.append('p')
	if action in drug_traces:
		found.append('d')
	return (found)


available_action = available_actions(initial_state)
action = sample_next_action(available_action)

env_matrix_snap = np.matrix(np.zeros([MATRIX_SIZE, MATRIX_SIZE]))

def update(current_state, action, gamma):
  max_index = np.where(Q[action, ] == np.max(Q[action, ]))[1]
  if max_index.shape[0] > 1:
	  max_index = int(np.random.choice(max_index, size = 1))
  else:
	  max_index = int(max_index)
  max_value = Q[action, max_index]
  Q[current_state, action] = M[current_state, action] + gamma * max_value
  environment = collect_environmental_data(action)
  if 'p' in environment:
  	env_police[current_state, action] += 1
  if 'd' in environment:
	  env_drugs[current_state, action] += 1
  if (np.max(Q) > 0):
	  return(np.sum(Q / np.max(Q)*100))
  else:
	  return (0)
# Same as above
update(initial_state, action, gamma)

def available_actions_with_env_help(state):
	current_state_row = M[state, ]
	av_action = np.where(current_state_row >= 0)[1]

	# if there are multiple routes, dis-favor anything negative
	env_pos_row = env_matrix_snap[state, av_action]

	if (np.sum(env_pos_row < 0)):
		# can we remove the negative directions from av_act?
		temp_av_action = av_action[np.array(env_pos_row)[0]>= 0]
		if len(temp_av_action) > 0:
			av_action = temp_av_action
	return av_action
# Determines the available actions according to the environment

```

# Different possible RL Architectures
RL architectures vary depending on how the components (states, actions, policies, and value functions) are represented and optimized

# 1. Tabular Architectures


*  Used for small state and action spaces.

*  Represent value functions or policies in lookup tables.
*  Inefficient for high-dimensional problems.






# 2. Neural Network Architectures




Neural networks are used to approximate value functions, policies, or both. Variations include:


*  Deep Q-Networks (DQN)
      

             1. Uses a CNN for state representation when states are images (e.g., Atari games)
             2. Two networks:
                Q-network (to predict Q(s,a)),Target network (for stable updates).







      

*   Policy Gradient Networks
             

             1.Directly parameterize the policy using a neural network.
             2.Outputs probabilities for each action in discrete action spaces or action values in continuous spaces.


*   Actor-Critic Architecture
           

             1.  Separate Networks: One for the actor (policy) and another for the critic (value function)
             2.  Shared Networks: A single shared backbone feeding both actor and critic heads.







* Recurrent Architectures
            



             1. Use RNNs, GRUs, or LSTMs to handle partially observable environments by maintaining a hidden state as memory.
             2. Example: Deep Recurrent Q-Network (DRQN)


* Transformer Architectures


             1.  Recently applied in RL to model long-term dependencies across sequences of states and actions.
             2. Example: Decision Transformers model RL tasks as sequence modeling problems.







# 3. Specialized Architectures

**Hierarchical RL:**

• Breaks tasks into subtasks, each with its own policy.

• Top-level policy selects subtasks, and lower-level policies execute actions.

• Example: Options framework.

**Multi-agent RL (MARL):**

• Handles environments with multiple interacting agents.

• Architectures for cooperation, competition, or mixed strategies.

• Examples: Independent Q-learning, MADDPG (Multi-Agent Deep Deterministic Policy Gradient).

**Meta-RL:**

• Enables agents to quickly adapt to new tasks by learning transferable knowledge.

• Example: Model-Agnostic Meta-Learning (MAML) for RL.

**Exploration-focused Architectures:**

• Encourage exploration by using intrinsic rewards.

• Example: Curiosity-driven RL uses a forward model to reward novelty.










# **Advanced RL Algorithms with Architectures**

**Soft Actor-Critic (SAC):**

• Uses stochastic policies and maximizes a trade-off between expected return and entropy (exploration).

• Uses two critics and a target entropy.

**Deep Deterministic Policy Gradient (DDPG):**

• Extends DQN to continuous action spaces by learning a deterministic policy.

**Trust Region Policy Optimization (TRPO):**

• Constrains policy updates to prevent large changes that could destabilize training.

**AlexNet**

**Description:** Revolutionized deep learning by winning ImageNet in 2012, using deeper layers and ReLU activation.

**Key Features:**
Overlapping max-pooling.
Dropout for regularization.

**Applications:** Large-scale image classification.

```python
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# Load the CIFAR-10 dataset
(train_data, val_data), info = tfds.load(
    'cifar10',
    split=['train[:80%]', 'train[80%:]'],
    with_info=True,
    as_supervised=True
)

# Resize images to match AlexNet's input size (227x227 pixels)
def preprocess_image(image, label):
    image = tf.image.resize(image, (227, 227))
    return image, label

train_data = train_data.map(preprocess_image)
val_data = val_data.map(preprocess_image)

# Define AlexNet-like architecture
def alexnet_like(input_shape=(227, 227, 3), num_classes=10):
    model = models.Sequential()

    # Adjust the layers for smaller input size
    model.add(layers.Conv2D(96, (11, 11), strides=(4, 4), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((3, 3), strides=(2, 2)))

    model.add(layers.Conv2D(256, (5, 5), activation='relu'))
    model.add(layers.MaxPooling2D((3, 3), strides=(2, 2)))

    model.add(layers.Conv2D(384, (3, 3), activation='relu'))
    model.add(layers.Conv2D(384, (3, 3), activation='relu'))
    model.add(layers.Conv2D(256, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((3, 3), strides=(2, 2)))

    model.add(layers.Flatten())
    model.add(layers.Dense(4096, activation='relu'))
    model.add(layers.Dropout(0.5))

    model.add(layers.Dense(4096, activation='relu'))
    model.add(layers.Dropout(0.5))

    model.add(layers.Dense(num_classes, activation='softmax'))

    return model

# Create an instance of the model
model = alexnet_like()

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Define batch size and number of epochs
batch_size = 128
epochs = 10

# Preprocess the data and enable shuffling
train_data = train_data.shuffle(buffer_size=50000).batch(batch_size).prefetch(buffer_size=tf.data.AUTOTUNE)
val_data = val_data.batch(batch_size).prefetch(buffer_size=tf.data.AUTOTUNE)

# Train the model
history = model.fit(train_data, epochs=epochs, validation_data=val_data)

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

**ResNet (Residual Networks)**

**Description:** Introduced residual connections to tackle the vanishing gradient problem in deep networks.

**Key Features:**
Skip connections bypass layers.
Enables training very deep networks (e.g., ResNet-50, ResNet-152).

**Applications:** Classification, object detection, segmentation.

```python
from keras.layers import Conv2D, BatchNormalization, Activation, Add, Input
from keras.models import Model

def resnet_block(input_tensor, filters):
    x = Conv2D(filters, (3, 3), padding='same')(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(filters, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Add()([x, input_tensor])
    x = Activation('relu')(x)
    return x

input_layer = Input(shape=(64, 64, 3))
x = resnet_block(input_layer, 64)
model = Model(inputs=input_layer, outputs=x)
```

**Sequence-to-Sequence RNN (Seq2Seq)**

**Description:** Encodes an input sequence into a fixed representation and decodes it into an output sequence. Popular for tasks with varying input/output lengths.

**Applications:**

Machine translation.

Text summarization.

```python
from keras.layers import Input, LSTM, Dense
from keras.models import Model

# Encoder
encoder_inputs = Input(shape=(None, 50))
encoder = LSTM(128, return_state=True)
encoder_outputs, state_h, state_c = encoder(encoder_inputs)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = Input(shape=(None, 50))
decoder_lstm = LSTM(128, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)
decoder_dense = Dense(50, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
```

**CTC (Connectionist Temporal Classification)**

CTC is not a type of recurrent unit like LSTM or GRU; rather, it’s a loss function designed for sequence-to-sequence problems where the input and output lengths may not be the same, and alignment between them is unknown. It is commonly used in tasks like speech recognition, handwriting recognition, or OCR.

**Why CTC?**

In tasks like speech recognition, the input (audio spectrogram) and the output (text transcription) have different lengths, and we don't know which input feature corresponds to which output character. CTC helps learn this mapping.

**How CTC Works**

**Blank Token:** CTC introduces a special blank token (-) to represent no output for a given input frame.

**Multiple Alignments:** CTC considers all possible alignments of input-output sequences and sums the probabilities.

**Decoding:**
Probable paths are reduced using rules like merging consecutive identical labels (aa-abb becomes aab), and blank tokens are removed.

**Transformer Architecture**

The Transformer architecture is a deep learning model primarily used in natural language processing (NLP) tasks. It was introduced in the paper "Attention is All You Need" by Vaswani et al. in 2017. The key innovation of the Transformer is its use of the self-attention mechanism, which allows the model to weigh the importance of different words in a sentence, regardless of their position.

**Key Components:**

**Self-Attention:**

Each word in the input sequence attends to every other word to capture dependencies, enabling the model to learn relationships between words in parallel rather than sequentially (as in RNNs).
Positional Encoding:

Since Transformers do not process data sequentially, positional encodings are added to the input embeddings to retain information about the order of the words.

**Encoder-Decoder Structure:**

**Encoder:** Processes the input sequence and generates a representation.

**Decoder:** Uses the encoder's output to generate the final output sequence.

**Multi-Head Attention:**

The self-attention mechanism is applied multiple times in parallel (hence "multi-head"), allowing the model to focus on different parts of the input simultaneously.

**Feed-Forward Networks:**

After the attention mechanism, the output passes through fully connected layers (feed-forward networks), followed by normalization and residual connections for stability.

**Attention Mechanism**

image.png

The Attention Mechanism enables models to focus on relevant parts of the input sequence when making predictions, improving performance on tasks with long-range dependencies.

**Key Components:**

**Query, Key, and Value (Q, K, V):**

**Query (Q):** The current word or token of interest.

**Key (K):** Each word/token in the sequence to compare against the query.

**Value (V):** Contains the information used for generating the output.
Attention Score:


A similarity score between the Query and Key is computed, typically using a dot product, scaled, and normalized via softmax.

**Weighted Sum:**

The attention scores are applied to the Values to compute a weighted sum, generating the output.

**Multi-Head Attention:**

Multiple attention mechanisms run in parallel, each focusing on different parts of the input, and their outputs are combined for a richer representation.

**Advantages:**
Captures Long-Range Dependencies: Focuses on any part of the sequence, overcoming RNN limitations.

**Parallelization:** Processes sequences in parallel, speeding up training.
Interpretability: Attention weights highlight which parts of the input are most important.

**Generative Adversarial Networks(GANs)**

A Generative Adversarial Network (GAN) is a class of machine learning models used for generating new data that resembles a given dataset. The architecture consists of two neural networks, the Generator and the Discriminator, which are trained together in a process known as adversarial training. The goal of a GAN is to generate data that is indistinguishable from real data, such as creating realistic images, audio, or text.

**Components of a GAN**

**Generator (G):**

The Generator creates fake data (e.g., images) from random noise (latent space).
Its goal is to produce data that looks as realistic as possible to fool the Discriminator into thinking it is real.

**Discriminator (D):**

The Discriminator distinguishes between real data (from the dataset) and fake data (generated by the Generator).

Its goal is to correctly classify whether an input is real or fake.

**Training Process:**

The Generator aims to generate data that fools the Discriminator into thinking it's real.

The Discriminator tries to correctly distinguish between real and fake data.

Both networks improve through a feedback loop: the Generator learns to create better fake data, while the Discriminator gets better at detecting fake data.

This process is a minimax game where the Generator minimizes the Discriminator's ability to detect fake data, while the Discriminator maximizes its ability to distinguish real from fake.

**Auto Encoders**

image.png

**An Autoencoder** is a type of artificial neural network used for unsupervised learning, primarily for data compression, denoising, and dimensionality reduction. It learns to encode input data into a smaller representation and then reconstruct it back to the original form.

**Components of an Autoencoder:**

**Encoder:**

The encoder compresses the input data into a latent space representation (a lower-dimensional code). It maps the input data into a hidden representation.

**Latent Space:**

The compressed code or bottleneck that holds the essential information from the input data, usually of lower dimensions.

**Decoder:**

The decoder reconstructs the data from the compressed representation, aiming to generate an approximation of the original input.

**How It Works:**

The Autoencoder is trained to minimize the reconstruction error between the input data and the output generated by the decoder.

The network learns to capture the most important features of the data, discarding unnecessary details to create a more compact representation.
